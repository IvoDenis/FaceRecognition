```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(24,12))
plt.rc('axes', labelsize=20)
plt.rc('axes',titlesize=20)
plt.rc('legend', fontsize=20)
plt.title("График истинности") # заголовок
plt.xlabel("X, пиксель") # ось абсцисс
plt.ylabel("Y, пиксель") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(df_1['Unnamed: 0'], df_1['result'],'r',label="Результат")  # построение графика
plt.legend(loc=0)
plt.savefig('График истинности')
```
